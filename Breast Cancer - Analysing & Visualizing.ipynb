```python
#'''Columns:
#. id
#.diagnosis : M - Malignant(cancerous), B - Benign(non-cancerous)
#.radius_mean : mean radius of the cell nuclei(a higher "radius_mean" value may be associated with larger cell nuclei)
#.texture_mean : variation in pixel intensity and patterns in the image(Texture features provide information about the
#variation in color and intensity in the images of the cell nuclei and their surrounding regions.)
#.perimeter_mean : mean perimeter of the cell nuclei(This feature can provide valuable information about the shape and structure
#of cell nuclei, which is useful for characterizing and classifying breast tumors
#as either malignant (cancerous) or benign (non-cancerous).)
#.area_mean : mean area of the cell nuclei( It provides information about the size and shape of cell nuclei.
#Larger cell nuclei are often associated with malignant (cancerous) tumors, 
#as cancer cells tend to have larger and more irregularly shaped nuclei. 
#Benign (non-cancerous) tumors typically have smaller, more regular cell nuclei.)
#.smoothness_mean : mean smoothness realated to cell nuclei which provides insights into the regularity or irregularity 
#of the cell boundaries. 
#.compactness_mean : mean compactness related to cell nuclei.( refers to how closely the shape of a cell nucleus resembles a
#circle.
#It is a measure of the compactness or roundness of the nucleus.cancerous : less compact or more irregular,
#benign : more round and compact )
#concavity_mean : measures the degree of inward or outward curvature of the cell nuclei's boundary.
#(Malignant cell nuclei tend to exhibit greater concavity and irregularities compared to benign cell nuclei,
#which are typically smoother and less concave.)
#.concave_points_mean : mean number of concave pints in the cell nuclei.(Malignant cell nuclei often exhibit more and deeper
#concave points compared to benign cell nuclei, which typically have fewer and shallower concave points.)
#.radius_worst : feature that provides information about the size of the largest cell nuclei in a tumor.
#.texture_worst :  This feature represents the worst or highest (i.e., the largest texture) measurements related to
#cell nuclei in breast tissue samples.
#.perimeter_worst : This feature represents the worst or largest (i.e., the largest perimeter) measurements
#related to cell nuclei in breast tissue samples..
#.area_worst : This feature represents the worst or largest (i.e., the largest area) measurements related to cell nuclei
#in breast tissue samples.
#.fractal_dimension_worst : mathematical measure that can describe the complexity & irregularity of shapes within the image.'''


```
